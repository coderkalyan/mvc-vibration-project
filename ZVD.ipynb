```python
from sympy import *
init_printing()
```

```python
A = IndexedBase('A')
t = IndexedBase('t')
zeta, omega = symbols("zeta omega")
omega_d = omega * sqrt(1 - zeta ** 2)
i = symbols("i", cls=Idx)
```

```python
C = Sum(A[i] * exp(zeta * omega * t[i]) * cos(omega_d * t[i]), [i, 1, 3])
S = Sum(A[i] * exp(zeta * omega * t[i]) * sin(omega_d * t[i]), [i, 1, 3])
```

```python
C
```

```python
S
```

```python
diff(C, omega)
```

```python
c_diff = diff(C, omega).doit()
c_diff.subs({t[1]: 0})
```

```python
s_diff = diff(S, omega).doit()
s_diff.subs({t[1]: 0})
```

```python
C = C.doit().subs({t[1]: 0, t[2]: pi / omega_d, t[3]: 2 * pi / omega_d})
S = S.doit().subs({t[1]: 0, t[2]: pi / omega_d, t[3]: 2 * pi / omega_d})
c_diff = c_diff.subs({t[1]: 0, t[2]: pi / omega_d, t[3]: 2 * pi / omega_d})
s_diff = s_diff.subs({t[1]: 0, t[2]: pi / omega_d, t[3]: 2 * pi / omega_d})
```

```python
s_diff
```

```python
E1 = Eq(C, 0)
E2 = Eq(S, 0)
E3 = Eq(c_diff, 0)
E4 = Eq(s_diff, 0)
E5 = Eq(A[1] + A[2] + A[3], 1)
```

```python
solution = solve([E1, E2, E3, E4, E5], [A[1], A[2], A[3]])
```
